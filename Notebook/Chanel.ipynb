Ce bloc télécharge chaque image du dataset, la redimensionne proprement au format 224×224 avec padding blanc puis la sauvegarde localement dans images_temp.
Il crée ensuite un tableau contenant le chemin, l’image prétraitée et les métadonnées associées (catégorie, titre…).
Au final, il génère un DataFrame df_images regroupant toutes les images valides téléchargées.

```python
# ============================================================================
# ÉTAPE 2 : TÉLÉCHARGEMENT ET PRÉTRAITEMENT DES IMAGES
# ============================================================================
print("\n[2/9] Téléchargement et prétraitement des images...")

os.makedirs('images_temp', exist_ok=True)

def download_and_preprocess_image(url, idx, target_size=(224, 224)):
    """Télécharge et prétraite une image"""
    try:
        response = requests.get(url, timeout=10)
        img = Image.open(BytesIO(response.content)).convert('RGB')

        # Redimensionnement
        img.thumbnail(target_size, Image.Resampling.LANCZOS)
        new_img = Image.new('RGB', target_size, (255, 255, 255))
        paste_x = (target_size[0] - img.width) // 2
        paste_y = (target_size[1] - img.height) // 2
        new_img.paste(img, (paste_x, paste_y))

        # Sauvegarde
        path = f'images_temp/img_{idx}.jpg'
        new_img.save(path)

        return path, new_img
    except:
        return None, None

images_data = []
for idx, row in tqdm(df_sample.iterrows(), total=len(df_sample), desc="Téléchargement"):
    if pd.notna(row.get('imageurl')):
        path, img = download_and_preprocess_image(row['imageurl'], idx)
        if path:
            images_data.append({
                'index': idx,
                'path': path,
                'image': img,
                'category': row.get('category2_code', 'Unknown'),
                'title': row.get('title', 'Unknown')
            })

df_images = pd.DataFrame(images_data)
print(f"\n✓ Images téléchargées: {len(df_images)}/{len(df_sample)}")
```

Ce bloc calcule pour chaque image un descripteur de texture en appliquant la méthode LBP (Local Binary Patterns), qui compare chaque pixel à ses voisins pour décrire les motifs locaux.
Les histogrammes résultants sont stockés dans un tableau lbp_features, représentant la signature texturale de chaque image du dataset.

```python
# ============================================================================
# ÉTAPE 3 : MÉTHODE CLASSIQUE 1 - LOCAL BINARY PATTERNS (LBP)
# ============================================================================
print("\n[3/9] Extraction des descripteurs LBP (Local Binary Patterns)...")

def extract_lbp_features(image, radius=3, n_points=24):
    """
    Extrait les caractéristiques LBP d'une image.
    LBP capture la texture locale en comparant chaque pixel à ses voisins.
    """
    # Conversion en niveaux de gris
    gray = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2GRAY)

    # Calcul LBP
    lbp = local_binary_pattern(gray, n_points, radius, method='uniform')

    # Histogramme des patterns (descripteur)
    n_bins = n_points + 2
    hist, _ = np.histogram(lbp.ravel(), bins=n_bins, range=(0, n_bins), density=True)

    return hist

lbp_features = []
for idx, row in tqdm(df_images.iterrows(), total=len(df_images), desc="LBP"):
    features = extract_lbp_features(row['image'])
    lbp_features.append(features)

lbp_features = np.array(lbp_features)
print(f"✓ Descripteurs LBP extraits: shape {lbp_features.shape}")
```

Ce bloc extrait pour chaque image un descripteur HOG, qui analyse l’orientation des gradients pour capturer la forme et les contours.
Les vecteurs HOG obtenus sont ensuite réunis dans hog_features, représentant la signature de structure et de silhouette de chaque image.

```python
# ============================================================================
# ÉTAPE 4 : MÉTHODE CLASSIQUE 2 - HOG (Histogram of Oriented Gradients)
# ============================================================================
print("\n[4/9] Extraction des descripteurs HOG...")

def extract_hog_features(image, orientations=9, pixels_per_cell=(16, 16), cells_per_block=(2, 2)):
    """
    Extrait les caractéristiques HOG d'une image.
    HOG capture la forme en analysant les gradients d'intensité.
    """
    # Conversion en niveaux de gris
    gray = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2GRAY)

    # Calcul HOG
    features = hog(
        gray,
        orientations=orientations,
        pixels_per_cell=pixels_per_cell,
        cells_per_block=cells_per_block,
        block_norm='L2-Hys',
        visualize=False,
        feature_vector=True
    )

    return features

hog_features = []
for idx, row in tqdm(df_images.iterrows(), total=len(df_images), desc="HOG"):
    features = extract_hog_features(row['image'])
    hog_features.append(features)

hog_features = np.array(hog_features)
print(f"✓ Descripteurs HOG extraits: shape {hog_features.shape}")
```

Ce bloc utilise le modèle pré-entraîné ResNet50 (sans la couche de classification) pour extraire des embeddings profonds représentant le contenu visuel de chaque image.
Les représentations obtenues sont stockées dans resnet_features, formant des vecteurs riches en informations sémantiques.

```python
# ============================================================================
# ÉTAPE 5 : MÉTHODE DEEP LEARNING 1 - ResNet50 (Pré-entraîné)
# ============================================================================
print("\n[5/9] Extraction des embeddings ResNet50...")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}")

# Chargement ResNet50 pré-entraîné
resnet_model = models.resnet50(pretrained=True)
# Retirer la dernière couche (classification)
resnet_model = nn.Sequential(*list(resnet_model.children())[:-1])
resnet_model = resnet_model.to(device)
resnet_model.eval()

# Transformations
resnet_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

def extract_resnet_features(image):
    """Extrait les embeddings ResNet50"""
    img_tensor = resnet_transform(image).unsqueeze(0).to(device)

    with torch.no_grad():
        features = resnet_model(img_tensor)

    return features.squeeze().cpu().numpy()

resnet_features = []
for idx, row in tqdm(df_images.iterrows(), total=len(df_images), desc="ResNet50"):
    features = extract_resnet_features(row['image'])
    resnet_features.append(features)

resnet_features = np.array(resnet_features)
print(f"✓ Embeddings ResNet50 extraits: shape {resnet_features.shape}")
```

Ce script extrait les embeddings visuels des images à l’aide du modèle CLIP, en les normalisant pour les utiliser ensuite dans des tâches de vision-language ou de recherche d’images similaires.

```python
# ============================================================================
# ÉTAPE 6 : MÉTHODE DEEP LEARNING 2 - CLIP (Vision-Language)
# ============================================================================
print("\n[6/9] Extraction des embeddings CLIP...")

clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
clip_model.eval()

def extract_clip_features(image):
    """Extrait les embeddings visuels CLIP"""
    inputs = clip_processor(images=image, return_tensors="pt").to(device)

    with torch.no_grad():
        features = clip_model.get_image_features(**inputs)

    # Normalisation
    features = features / features.norm(dim=-1, keepdim=True)
    return features.squeeze().cpu().numpy()

clip_features = []
for idx, row in tqdm(df_images.iterrows(), total=len(df_images), desc="CLIP"):
    features = extract_clip_features(row['image'])
    clip_features.append(features)

clip_features = np.array(clip_features)
print(f"✓ Embeddings CLIP extraits: shape {clip_features.shape}")
```

Ce code extrait les embeddings des images avec le Vision Transformer (ViT) en utilisant le token CLS pour représenter chaque image, afin de les utiliser ensuite dans des tâches de vision ou d’analyse de similarité.

```python
# ============================================================================
# ÉTAPE 7 : MÉTHODE DEEP LEARNING 3 - ViT (Vision Transformer)
# ============================================================================
print("\n[7/9] Extraction des embeddings ViT (Vision Transformer)...")

vit_processor = ViTImageProcessor.from_pretrained('google/vit-base-patch16-224')
vit_model = ViTModel.from_pretrained('google/vit-base-patch16-224').to(device)
vit_model.eval()

def extract_vit_features(image):
    """Extrait les embeddings Vision Transformer"""
    inputs = vit_processor(images=image, return_tensors="pt").to(device)

    with torch.no_grad():
        outputs = vit_model(**inputs)
        # Utiliser le token CLS (classification)
        features = outputs.last_hidden_state[:, 0, :]

    return features.squeeze().cpu().numpy()

vit_features = []
for idx, row in tqdm(df_images.iterrows(), total=len(df_images), desc="ViT"):
    features = extract_vit_features(row['image'])
    vit_features.append(features)

vit_features = np.array(vit_features)
print(f"✓ Embeddings ViT extraits: shape {vit_features.shape}")
```

Ce code effectue une analyse comparative des différentes méthodes d’extraction de caractéristiques :

1 - Cohérence intra-classe : mesure à quel point les produits de la même catégorie sont proches dans l’espace des embeddings (plus faible = meilleure cohérence).

2 - Distance inter-classe : mesure la séparation moyenne entre produits de catégories différentes (plus élevé = meilleure séparation).

3 - Ratio inter/intra : combine les deux métriques pour évaluer globalement la qualité des embeddings (plus élevé = meilleur).

4 - Exemples représentatifs : calcule les distances cosinus entre 3 produits spécifiques pour comparer visuellement les méthodes.

```python
# ============================================================================
# ÉTAPE 8 : ANALYSE COMPARATIVE DES MÉTHODES
# ============================================================================
print("\n[8/9] Analyse comparative des méthodes...")

# Dictionnaire de toutes les méthodes
methods = {
    'LBP (Texture)': lbp_features,
    'HOG (Forme)': hog_features,
    'ResNet50': resnet_features,
    'CLIP': clip_features,
    'ViT': vit_features
}

# 8.1 - Cohérence intra-classe (produits similaires proches)
print("\n📊 8.1 - Cohérence intra-classe par catégorie...")

def compute_intra_class_coherence(features, categories):
    """
    Calcule la cohérence intra-classe : distance moyenne entre produits de même catégorie.
    Plus la valeur est FAIBLE, meilleure est la cohérence.
    """
    unique_categories = np.unique(categories)  # CORRIGÉ : np.unique() au lieu de .unique()
    coherence_scores = []

    for cat in unique_categories:
        cat_indices = np.where(categories == cat)[0]
        if len(cat_indices) < 2:
            continue

        cat_features = features[cat_indices]
        # Distance moyenne intra-classe
        distances = []
        for i in range(len(cat_features)):
            for j in range(i+1, len(cat_features)):
                dist = cosine(cat_features[i], cat_features[j])
                distances.append(dist)

        if distances:
            coherence_scores.append(np.mean(distances))

    return np.mean(coherence_scores) if coherence_scores else 0

categories = df_images['category'].values
intra_class_results = {}

for method_name, features in methods.items():
    score = compute_intra_class_coherence(features, categories)
    intra_class_results[method_name] = score
    print(f"  {method_name:20s}: {score:.4f}")

# 8.2 - Distance inter-classe (catégories différentes éloignées)
print("\n📊 8.2 - Distance inter-classe...")

def compute_inter_class_distance(features, categories):
    """
    Calcule la distance inter-classe : distance moyenne entre produits de catégories différentes.
    Plus la valeur est ÉLEVÉE, meilleure est la séparation.
    """
    unique_categories = np.unique(categories)  # CORRIGÉ : np.unique() au lieu de .unique()
    if len(unique_categories) < 2:
        return 0

    distances = []
    for i, cat1 in enumerate(unique_categories):
        for cat2 in unique_categories[i+1:]:
            cat1_indices = np.where(categories == cat1)[0]
            cat2_indices = np.where(categories == cat2)[0]

            cat1_features = features[cat1_indices]
            cat2_features = features[cat2_indices]

            for f1 in cat1_features[:5]:  # Limiter pour performance
                for f2 in cat2_features[:5]:
                    dist = cosine(f1, f2)
                    distances.append(dist)

    return np.mean(distances) if distances else 0

inter_class_results = {}

for method_name, features in methods.items():
    score = compute_inter_class_distance(features, categories)
    inter_class_results[method_name] = score
    print(f"  {method_name:20s}: {score:.4f}")

# 8.3 - Ratio de séparation (inter/intra)
print("\n📊 8.3 - Ratio de séparation (inter/intra) - PLUS ÉLEVÉ = MEILLEUR...")

separation_ratios = {}
for method_name in methods.keys():
    if intra_class_results[method_name] > 0:
        ratio = inter_class_results[method_name] / intra_class_results[method_name]
    else:
        ratio = 0
    separation_ratios[method_name] = ratio
    print(f"  {method_name:20s}: {ratio:.4f}")

# 8.4 - Exemples représentatifs : Comparaison sur 3 produits spécifiques
print("\n📊 8.4 - Analyse sur exemples représentatifs...")

# Sélection de 3 produits de catégories différentes
sample_categories = df_images['category'].value_counts().head(3).index.tolist()
sample_products = []
for cat in sample_categories:
    prod = df_images[df_images['category'] == cat].iloc[0]
    sample_products.append(prod)

print("\nProduits sélectionnés:")
for i, prod in enumerate(sample_products):
    print(f"  {i+1}. {prod['title'][:50]}... ({prod['category']})")

# Calcul des distances entre ces 3 produits pour chaque méthode
print("\nDistances entre produits (cosine):")
for method_name, features in methods.items():
    print(f"\n  {method_name}:")
    sample_indices = [prod['index'] for prod in sample_products]

    for i in range(len(sample_products)):
        for j in range(i+1, len(sample_products)):
            idx_i = df_images[df_images['index'] == sample_indices[i]].index[0]
            idx_j = df_images[df_images['index'] == sample_indices[j]].index[0]

            dist = cosine(features[idx_i], features[idx_j])
            print(f"    Produit {i+1} ↔ Produit {j+1}: {dist:.4f}")
```

Ce code met en place un pipeline complet d’extraction de descripteurs textuels à partir des titres des produits. Il inclut des méthodes classiques comme Bag of Words et TF-IDF, des embeddings de mots via Word2Vec et FastText, ainsi que des modèles de type Transformers (BERT, DistilBERT, CLIP Text) pour obtenir des représentations vectorielles plus riches. Ces descripteurs permettent de comparer les produits, analyser les similarités intra- et inter-catégories, et visualiser les relations entre produits dans un espace réduit grâce à UMAP ou t-SNE. L’approche est flexible et peut être utilisée pour l’analyse exploratoire, la recherche de doublons ou des tâches avancées de machine learning.

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.manifold import TSNE
import umap
import warnings
warnings.filterwarnings('ignore')

# Deep Learning / Transformers
import torch
from transformers import (
    BertTokenizer, BertModel,
    DistilBertTokenizer, DistilBertModel,
    CLIPTokenizer, CLIPTextModel
)
from gensim.models import Word2Vec, FastText
from gensim.utils import simple_preprocess
import re
from collections import Counter

# Configuration
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
torch.manual_seed(RANDOM_STATE)

class TextDescriptorExtractor:
    """Classe principale pour l'extraction de descripteurs textuels"""

    def __init__(self, df, text_column='title'):
        self.df = df.copy()
        self.text_column = text_column
        self.texts = df[text_column].fillna('').astype(str).tolist()
        self.cleaned_texts = self._clean_texts()
        self.descriptors = {}

    def _clean_texts(self):
        """Nettoyage et prétraitement des textes"""
        cleaned = []
        for text in self.texts:
            # Convertir en minuscules
            text = text.lower()
            # Supprimer les caractères spéciaux mais garder les espaces
            text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
            # Supprimer les espaces multiples
            text = re.sub(r'\s+', ' ', text).strip()
            cleaned.append(text)
        return cleaned

    # ==================== MÉTHODES CLASSIQUES ====================

    def extract_bow(self, max_features=500):
        """Bag of Words"""
        print("🔤 Extraction BoW...")
        vectorizer = CountVectorizer(
            max_features=max_features,
            stop_words='english',
            min_df=2
        )
        bow_matrix = vectorizer.fit_transform(self.cleaned_texts)
        self.descriptors['bow'] = bow_matrix.toarray()
        print(f"   Shape: {self.descriptors['bow'].shape}")
        return self.descriptors['bow'], vectorizer

    def extract_tfidf(self, max_features=500):
        """TF-IDF"""
        print("📊 Extraction TF-IDF...")
        vectorizer = TfidfVectorizer(
            max_features=max_features,
            stop_words='english',
            min_df=2,
            ngram_range=(1, 2)  # Unigrams et bigrams
        )
        tfidf_matrix = vectorizer.fit_transform(self.cleaned_texts)
        self.descriptors['tfidf'] = tfidf_matrix.toarray()
        print(f"   Shape: {self.descriptors['tfidf'].shape}")
        return self.descriptors['tfidf'], vectorizer

    # ==================== WORD EMBEDDINGS ====================

    def extract_word2vec(self, vector_size=100, window=5, min_count=2):
        """Word2Vec embeddings"""
        print("🧠 Entraînement Word2Vec...")
        tokenized_texts = [simple_preprocess(text) for text in self.cleaned_texts]

        model = Word2Vec(
            sentences=tokenized_texts,
            vector_size=vector_size,
            window=window,
            min_count=min_count,
            workers=4,
            seed=RANDOM_STATE
        )

        # Moyenne des vecteurs de mots pour chaque document
        embeddings = []
        for tokens in tokenized_texts:
            vectors = [model.wv[word] for word in tokens if word in model.wv]
            if vectors:
                embeddings.append(np.mean(vectors, axis=0))
            else:
                embeddings.append(np.zeros(vector_size))

        self.descriptors['word2vec'] = np.array(embeddings)
        print(f"   Shape: {self.descriptors['word2vec'].shape}")
        return self.descriptors['word2vec'], model

    def extract_fasttext(self, vector_size=100, window=5, min_count=2):
        """FastText embeddings"""
        print("⚡ Entraînement FastText...")
        tokenized_texts = [simple_preprocess(text) for text in self.cleaned_texts]

        model = FastText(
            sentences=tokenized_texts,
            vector_size=vector_size,
            window=window,
            min_count=min_count,
            workers=4,
            seed=RANDOM_STATE
        )

        embeddings = []
        for tokens in tokenized_texts:
            vectors = [model.wv[word] for word in tokens if word in model.wv]
            if vectors:
                embeddings.append(np.mean(vectors, axis=0))
            else:
                embeddings.append(np.zeros(vector_size))

        self.descriptors['fasttext'] = np.array(embeddings)
        print(f"   Shape: {self.descriptors['fasttext'].shape}")
        return self.descriptors['fasttext'], model

    # ==================== TRANSFORMERS ====================

    def extract_bert(self, model_name='bert-base-uncased', batch_size=16):
        """BERT embeddings"""
        print(f"🤖 Extraction BERT ({model_name})...")
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

        tokenizer = BertTokenizer.from_pretrained(model_name)
        model = BertModel.from_pretrained(model_name).to(device)
        model.eval()

        embeddings = []
        with torch.no_grad():
            for i in range(0, len(self.cleaned_texts), batch_size):
                batch_texts = self.cleaned_texts[i:i+batch_size]

                inputs = tokenizer(
                    batch_texts,
                    padding=True,
                    truncation=True,
                    max_length=128,
                    return_tensors='pt'
                ).to(device)

                outputs = model(**inputs)
                # Utiliser le [CLS] token
                batch_embeddings = outputs.last_hidden_state[:, 0, :].cpu().numpy()
                embeddings.extend(batch_embeddings)

        self.descriptors['bert'] = np.array(embeddings)
        print(f"   Shape: {self.descriptors['bert'].shape}")
        return self.descriptors['bert']

    def extract_distilbert(self, batch_size=16):
        """DistilBERT embeddings (plus rapide que BERT)"""
        print("🚀 Extraction DistilBERT...")
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

        tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
        model = DistilBertModel.from_pretrained('distilbert-base-uncased').to(device)
        model.eval()

        embeddings = []
        with torch.no_grad():
            for i in range(0, len(self.cleaned_texts), batch_size):
                batch_texts = self.cleaned_texts[i:i+batch_size]

                inputs = tokenizer(
                    batch_texts,
                    padding=True,
                    truncation=True,
                    max_length=128,
                    return_tensors='pt'
                ).to(device)

                outputs = model(**inputs)
                batch_embeddings = outputs.last_hidden_state[:, 0, :].cpu().numpy()
                embeddings.extend(batch_embeddings)

        self.descriptors['distilbert'] = np.array(embeddings)
        print(f"   Shape: {self.descriptors['distilbert'].shape}")
        return self.descriptors['distilbert']

    def extract_clip_text(self, batch_size=16):
        """CLIP Text Encoder"""
        print("🎨 Extraction CLIP Text...")
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

        tokenizer = CLIPTokenizer.from_pretrained('openai/clip-vit-base-patch32')
        model = CLIPTextModel.from_pretrained('openai/clip-vit-base-patch32').to(device)
        model.eval()

        embeddings = []
        with torch.no_grad():
            for i in range(0, len(self.cleaned_texts), batch_size):
                batch_texts = self.cleaned_texts[i:i+batch_size]

                inputs = tokenizer(
                    batch_texts,
                    padding=True,
                    truncation=True,
                    max_length=77,
                    return_tensors='pt'
                ).to(device)

                outputs = model(**inputs)
                batch_embeddings = outputs.pooler_output.cpu().numpy()
                embeddings.extend(batch_embeddings)

        self.descriptors['clip_text'] = np.array(embeddings)
        print(f"   Shape: {self.descriptors['clip_text'].shape}")
        return self.descriptors['clip_text']


class TextDescriptorAnalyzer:
    """Analyse comparative des descripteurs textuels"""

    def __init__(self, extractor):
        self.extractor = extractor
        self.df = extractor.df

    def compute_similarities(self, descriptor_name, sample_indices=None):
        """Calcule la matrice de similarité cosinus"""
        if descriptor_name not in self.extractor.descriptors:
            raise ValueError(f"Descripteur '{descriptor_name}' non trouvé")

        descriptors = self.extractor.descriptors[descriptor_name]

        if sample_indices is not None:
            descriptors = descriptors[sample_indices]

        return cosine_similarity(descriptors)

    def intra_inter_class_analysis(self, descriptor_name, category_col='category2_code'):
        """Analyse intra-classe vs inter-classe"""
        print(f"\n📊 Analyse intra/inter-classe pour {descriptor_name}")

        descriptors = self.extractor.descriptors[descriptor_name]
        categories = self.df[category_col].values

        intra_similarities = []
        inter_similarities = []

        # Échantillonner pour accélérer
        sample_size = min(500, len(descriptors))
        indices = np.random.choice(len(descriptors), sample_size, replace=False)

        for i in indices:
            for j in indices:
                if i >= j:
                    continue

                sim = cosine_similarity(
                    descriptors[i].reshape(1, -1),
                    descriptors[j].reshape(1, -1)
                )[0, 0]

                if categories[i] == categories[j]:
                    intra_similarities.append(sim)
                else:
                    inter_similarities.append(sim)

        print(f"   Intra-classe (même catégorie) : {np.mean(intra_similarities):.4f} ± {np.std(intra_similarities):.4f}")
        print(f"   Inter-classe (différentes)    : {np.mean(inter_similarities):.4f} ± {np.std(inter_similarities):.4f}")
        print(f"   Ratio intra/inter             : {np.mean(intra_similarities) / np.mean(inter_similarities):.4f}")

        return intra_similarities, inter_similarities

    def visualize_embeddings(self, descriptor_name, method='umap', n_samples=1000):
        """Visualisation 2D des embeddings"""
        print(f"\n🎨 Visualisation {method.upper()} pour {descriptor_name}")

        descriptors = self.extractor.descriptors[descriptor_name]

        # Échantillonner si trop grand
        if len(descriptors) > n_samples:
            indices = np.random.choice(len(descriptors), n_samples, replace=False)
            descriptors_sample = descriptors[indices]
            df_sample = self.df.iloc[indices]
        else:
            descriptors_sample = descriptors
            df_sample = self.df

        # Réduction de dimensionnalité
        if method == 'tsne':
            reducer = TSNE(n_components=2, random_state=RANDOM_STATE, perplexity=30)
        else:  # umap
            reducer = umap.UMAP(n_components=2, random_state=RANDOM_STATE, n_neighbors=15)

        embeddings_2d = reducer.fit_transform(descriptors_sample)

        # Visualisation
        plt.figure(figsize=(12, 8))
        categories = df_sample['category2_code'].values
        unique_cats = np.unique(categories)

        for cat in unique_cats:
            mask = categories == cat
            plt.scatter(
                embeddings_2d[mask, 0],
                embeddings_2d[mask, 1],
                label=cat,
                alpha=0.6,
                s=50
            )

        plt.title(f'{method.upper()} - {descriptor_name}', fontsize=14, fontweight='bold')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
        plt.xlabel('Dimension 1')
        plt.ylabel('Dimension 2')
        plt.tight_layout()
        plt.show()

    def compare_methods_visualization(self, top_methods=4):
        """Compare visuellement plusieurs méthodes"""
        methods = list(self.extractor.descriptors.keys())[:top_methods]

        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        axes = axes.flatten()

        for idx, method in enumerate(methods):
            descriptors = self.extractor.descriptors[method]

            # UMAP réduction
            n_samples = min(800, len(descriptors))
            indices = np.random.choice(len(descriptors), n_samples, replace=False)
            descriptors_sample = descriptors[indices]
            df_sample = self.df.iloc[indices]

            reducer = umap.UMAP(n_components=2, random_state=RANDOM_STATE)
            embeddings_2d = reducer.fit_transform(descriptors_sample)

            # Plot
            categories = df_sample['category2_code'].values
            for cat in np.unique(categories):
                mask = categories == cat
                axes[idx].scatter(
                    embeddings_2d[mask, 0],
                    embeddings_2d[mask, 1],
                    label=cat,
                    alpha=0.6,
                    s=30
                )

            axes[idx].set_title(method.upper(), fontsize=12, fontweight='bold')
            axes[idx].legend(fontsize=7)

        plt.suptitle('Comparaison des méthodes - UMAP', fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.show()

    def similarity_heatmap(self, descriptor_name, n_samples=50):
        """Heatmap de similarité"""
        descriptors = self.extractor.descriptors[descriptor_name]

        indices = np.random.choice(len(descriptors), n_samples, replace=False)
        descriptors_sample = descriptors[indices]

        sim_matrix = cosine_similarity(descriptors_sample)

        plt.figure(figsize=(10, 8))
        sns.heatmap(sim_matrix, cmap='YlOrRd', square=True, cbar_kws={'label': 'Cosine Similarity'})
        plt.title(f'Matrice de Similarité - {descriptor_name}', fontsize=14, fontweight='bold')
        plt.xlabel('Produit index')
        plt.ylabel('Produit index')
        plt.tight_layout()
        plt.show()


# ==================== EXEMPLE D'UTILISATION ====================

if __name__ == "__main__":
    # Charger les données
    print("📦 Chargement des données...")
    df = pd.read_csv('chanel_products.csv')  # Adapter le chemin
    print(f"   Dataset: {len(df)} produits")

    # Initialiser l'extracteur
    extractor = TextDescriptorExtractor(df, text_column='title')

    # ========== EXTRACTION ==========
    print("\n" + "="*60)
    print("EXTRACTION DES DESCRIPTEURS TEXTUELS")
    print("="*60)

    # Méthodes classiques
    bow_features, bow_vectorizer = extractor.extract_bow(max_features=500)
    tfidf_features, tfidf_vectorizer = extractor.extract_tfidf(max_features=500)

    # Word embeddings
    w2v_features, w2v_model = extractor.extract_word2vec(vector_size=100)
    ft_features, ft_model = extractor.extract_fasttext(vector_size=100)

    # Transformers
    bert_features = extractor.extract_bert()
    distilbert_features = extractor.extract_distilbert()
    clip_features = extractor.extract_clip_text()

```

```python
# ============================================================================
# SYSTÈME DE RECOMMANDATION CHANEL -
# ============================================================================

# ============================================================================
# CELLULE 1 : Installation
# ============================================================================
print("📦 Installation des packages...")

!pip install -q streamlit pyngrok transformers torch datasets pillow scikit-learn deep-translator langdetect

print("✅ Installation terminée!")

# ============================================================================
# CELLULE 2 : Configuration Ngrok
# ============================================================================
from pyngrok import ngrok, conf
import getpass

print("🔧 Configuration de ngrok")
print("⚠️ Token : https://dashboard.ngrok.com/get-started/your-authtoken")

ngrok_token = getpass.getpass('Token ngrok: ')
conf.get_default().auth_token = ngrok_token

print("✅ Ngrok configuré!")

# ============================================================================
# CELLULE 3 : Chargement Dataset
# ============================================================================
from datasets import load_dataset
import pandas as pd

print("📥 Chargement dataset...")

dataset = load_dataset("DBQ/Chanel.Product.prices.Germany")
df = pd.DataFrame(dataset['train'])
df = df.dropna(subset=['title']).reset_index(drop=True)
df = df.head(1428)

print(f"✅ {len(df)} produits chargés")

!mkdir -p /content/data
df.to_pickle('/content/data/products_df.pkl')

# ============================================================================
# CELLULE 4 : Embeddings Visuels CLIP
# ============================================================================
import torch
import numpy as np
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import requests
from io import BytesIO
from tqdm import tqdm
import pickle

print("🎨 Génération embeddings visuels CLIP...")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}")

clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

visual_embeddings = []

for idx, row in tqdm(df.iterrows(), total=len(df), desc="CLIP"):
    try:
        if pd.notna(row.get('imageurl')):
            response = requests.get(row['imageurl'], timeout=10)
            image = Image.open(BytesIO(response.content)).convert('RGB')
            inputs = clip_processor(images=image, return_tensors="pt").to(device)

            with torch.no_grad():
                features = clip_model.get_image_features(**inputs)

            features = features / features.norm(dim=-1, keepdim=True)
            visual_embeddings.append(features.cpu().numpy().flatten())
        else:
            visual_embeddings.append(np.zeros(512))
    except:
        visual_embeddings.append(np.zeros(512))

visual_embeddings = np.array(visual_embeddings)

with open('/content/data/visual_embeddings.pkl', 'wb') as f:
    pickle.dump(visual_embeddings, f)

print(f"✅ Shape: {visual_embeddings.shape}")

# ============================================================================
# CELLULE 5 : Embeddings Textuels BERT
# ============================================================================
from transformers import BertTokenizer, BertModel

print("📝 Génération embeddings textuels BERT...")

tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
bert_model = BertModel.from_pretrained('bert-base-multilingual-cased').to(device)
bert_model.eval()

text_embeddings = []

for idx, row in tqdm(df.iterrows(), total=len(df), desc="BERT"):
    try:
        text = str(row.get('title', ''))
        inputs = tokenizer(text, return_tensors='pt', padding=True,
                          truncation=True, max_length=128).to(device)

        with torch.no_grad():
            outputs = bert_model(**inputs)

        embedding = outputs.last_hidden_state[:, 0, :].cpu().numpy().flatten()
        text_embeddings.append(embedding)
    except:
        text_embeddings.append(np.zeros(768))

text_embeddings = np.array(text_embeddings)

with open('/content/data/text_embeddings.pkl', 'wb') as f:
    pickle.dump(text_embeddings, f)

print(f"✅ Shape: {text_embeddings.shape}")

# ============================================================================
# CELLULE 6 : Application Streamlit OPTIMISÉE
# ============================================================================

app_code = """
import streamlit as st
import numpy as np
import pandas as pd
from PIL import Image
import pickle
from sklearn.metrics.pairwise import cosine_similarity
import requests
from io import BytesIO
from deep_translator import GoogleTranslator
from langdetect import detect
import torch
from transformers import CLIPProcessor, CLIPModel, BertTokenizer, BertModel

st.set_page_config(page_title="Chanel Recommender", page_icon="💄", layout="wide")

st.markdown('''
<style>
.main {background-color: #f8f9fa;}
.stButton>button {
    background-color: #000000;
    color: white;
    font-weight: bold;
    border-radius: 10px;
    padding: 10px 25px;
}
.stButton>button:hover {background-color: #333333;}
.similarity-score {
    background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
    color: white;
    padding: 5px 15px;
    border-radius: 20px;
    font-weight: bold;
    display: inline-block;
}
.category-badge {
    background: #4CAF50;
    color: white;
    padding: 3px 10px;
    border-radius: 12px;
    font-size: 12px;
    font-weight: bold;
}
</style>
''', unsafe_allow_html=True)

# ============================================================================
# MAPPING CATÉGORIES
# ============================================================================

CATEGORY_MAPPING = {
    # LUNETTES
    'lunettes': ['SUNGLASSES', 'EYEGLASSES', 'BLUE LIGHT GLASSES'],
    'lunette': ['SUNGLASSES', 'EYEGLASSES', 'BLUE LIGHT GLASSES'],
    'soleil': ['SUNGLASSES'],
    'glasses': ['SUNGLASSES', 'EYEGLASSES', 'BLUE LIGHT GLASSES'],
    'sunglasses': ['SUNGLASSES'],
    'eyeglasses': ['EYEGLASSES'],

    # SACS
    'sac': ['HANDBAGS'],
    'bag': ['HANDBAGS'],
    'handbag': ['HANDBAGS'],
    'sac à main': ['HANDBAGS'],
    'pochette': ['SMALL LEATHER GOODS'],
    'portefeuille': ['SMALL LEATHER GOODS'],

    # CHAUSSURES
    'chaussure': ['SHOES'],
    'chaussures': ['SHOES'],
    'shoe': ['SHOES'],
    'shoes': ['SHOES'],
    'escarpin': ['SHOES'],
    'botte': ['SHOES'],
    'sneaker': ['SHOES'],

    # VÊTEMENTS
    'vêtement': ['READY-TO-WEAR'],
    'vetement': ['READY-TO-WEAR'],
    'robe': ['READY-TO-WEAR'],
    'pantalon': ['READY-TO-WEAR'],
    'veste': ['READY-TO-WEAR'],
    'manteau': ['READY-TO-WEAR'],
    'pull': ['READY-TO-WEAR'],
    'chemise': ['READY-TO-WEAR'],

    # BIJOUX
    'bijou': ['JEWELRY', 'COSTUME JEWELRY', 'COCO CRUSH'],
    'bijoux': ['JEWELRY', 'COSTUME JEWELRY', 'COCO CRUSH'],
    'bague': ['RINGS'],
    'ring': ['RINGS'],
    'collier': ['NECKLACES'],
    'necklace': ['NECKLACES'],
    'boucle': ['EARRINGS'],
    'earring': ['EARRINGS'],
    'bracelet': ['BRACELETS'],
    'diamant': ['JEWELRY'],
    'or': ['JEWELRY'],

    # MONTRES
    'montre': ['WATCHES'],
    'watch': ['WATCHES'],

    # MAQUILLAGE
    'rouge à lèvres': ['LIPS'],
    'rouge a levres': ['LIPS'],
    'lipstick': ['LIPS'],
    'lips': ['LIPS'],
    'mascara': ['EYES'],
    'eye': ['EYES'],
    'yeux': ['EYES'],
    'fard': ['EYES'],
    'fond de teint': ['FACE'],
    'foundation': ['FACE'],
    'face': ['FACE'],
    'nail': ['NAILS'],
    'vernis': ['NAILS'],
    'ongle': ['NAILS'],
    'pinceau': ['BRUSHES AND ACCESSORIES'],
    'brush': ['BRUSHES AND ACCESSORIES'],

    # PARFUMS
    'parfum': ['FRAGRANCE', 'WOMEN', 'MEN'],
    'perfume': ['FRAGRANCE', 'WOMEN', 'MEN'],
    'eau de toilette': ['FRAGRANCE'],
    'cologne': ['MEN'],
    'fragrance': ['FRAGRANCE'],

    # ACCESSOIRES
    'ceinture': ['BELTS'],
    'belt': ['BELTS'],
    'chapeau': ['HEADWEAR'],
    'hat': ['HEADWEAR'],
    'écharpe': ['SCARVES'],
    'scarf': ['SCARVES'],
    'foulard': ['SCARVES'],

    # SKINCARE
    'crème': ['MOISTURIZERS', 'SKINCARE'],
    'creme': ['MOISTURIZERS', 'SKINCARE'],
    'moisturizer': ['MOISTURIZERS'],
    'sérum': ['SERUMS'],
    'serum': ['SERUMS'],
    'nettoyant': ['CLEANSERS & MAKEUP REMOVERS'],
    'cleanser': ['CLEANSERS & MAKEUP REMOVERS'],
    'masque': ['MASKS & EXFOLIATORS'],
    'mask': ['MASKS & EXFOLIATORS'],

    # FACE
    'teint': ['FACE'],
    'poudre': ['FACE'],
    'blush': ['FACE'],
    'correcteur': ['FACE'],
    'concealer': ['FACE'],
    'highlighter': ['FACE'],

    # BATH AND BODY
    'gel douche': ['BATH AND BODY'],
    'gel lavant': ['BATH AND BODY'],
    'savon': ['BATH AND BODY'],
    'lotion corps': ['BATH AND BODY'],
    'crème corps': ['BATH AND BODY'],
    'body lotion': ['BATH AND BODY'],
    'body wash': ['BATH AND BODY'],

    # EYES
    'eyeliner': ['EYES'],
    'crayon yeux': ['EYES'],
    'kohl': ['EYES'],
    'eyeshadow': ['EYES'],
    'ombre à paupières': ['EYES'],

    # MOISTURIZERS
    'hydratant': ['MOISTURIZERS'],
    'hydrating': ['MOISTURIZERS'],
    'day cream': ['MOISTURIZERS'],
    'night cream': ['MOISTURIZERS'],
}

def detect_category_from_query(query_text):
    query_lower = query_text.lower()
    for keyword, categories in CATEGORY_MAPPING.items():
        if keyword in query_lower:
            return categories
    return None

# ============================================================================
# TRADUCTION
# ============================================================================

class SmartTranslator:
    def __init__(self):
        self.translator = GoogleTranslator(source='auto', target='de')
        self.cache = {}

    def translate_to_german(self, text):
        if not text or text in self.cache:
            return self.cache.get(text, text)

        try:
            lang = detect(text)
            if lang == 'de':
                self.cache[text] = text
                return text

            translated = self.translator.translate(text)
            self.cache[text] = translated
            return translated
        except:
            return text

@st.cache_resource
def get_translator():
    return SmartTranslator()

translator = get_translator()

# ============================================================================
# CHARGEMENT DONNÉES
# ============================================================================

@st.cache_resource
def load_data():
    df = pd.read_pickle('/content/data/products_df.pkl')
    with open('/content/data/visual_embeddings.pkl', 'rb') as f:
        visual_emb = pickle.load(f)
    with open('/content/data/text_embeddings.pkl', 'rb') as f:
        text_emb = pickle.load(f)
    return df, visual_emb, text_emb

df, visual_emb, text_emb = load_data()

# ============================================================================
# MODÈLES CLIP ET BERT
# ============================================================================

@st.cache_resource
def load_clip_model():
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
    processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
    model.eval()
    return model, processor, device

@st.cache_resource
def load_bert_model():
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
    model = BertModel.from_pretrained('bert-base-multilingual-cased').to(device)
    model.eval()
    return tokenizer, model, device

clip_model, clip_processor, clip_device = load_clip_model()
tokenizer, bert_model, bert_device = load_bert_model()

# ============================================================================
# EXTRACTION EMBEDDINGS OPTIMISÉE
# ============================================================================

def extract_image_embedding(image):
    '''Extrait embedding CLIP d'une image uploadée avec normalisation optimale'''
    try:
        image_rgb = image.convert('RGB')
        inputs = clip_processor(images=image_rgb, return_tensors="pt").to(clip_device)

        with torch.no_grad():
            features = clip_model.get_image_features(**inputs)

        features = features / features.norm(dim=-1, keepdim=True)
        return features.cpu().numpy().flatten()
    except Exception as e:
        st.error(f"Erreur extraction image: {e}")
        return np.zeros(512)

def extract_text_embedding(text):
    try:
        inputs = tokenizer(text, return_tensors='pt', padding=True,
                          truncation=True, max_length=128).to(bert_device)

        with torch.no_grad():
            outputs = bert_model(**inputs)

        embedding = outputs.last_hidden_state[:, 0, :].cpu().numpy().flatten()
        return embedding / np.linalg.norm(embedding)
    except:
        return np.zeros(768)

# ============================================================================
# DÉTECTION CATÉGORIE PAR IMAGE avec CLIP
# ============================================================================

def detect_category_from_image(image):
    '''
    Utilise CLIP pour identifier la catégorie du produit dans l'image.
    Retourne la catégorie détectée et la confiance.
    '''
    category_labels = {
        'watch': ['WATCHES'],
        'sunglasses': ['SUNGLASSES', 'EYEGLASSES', 'BLUE LIGHT GLASSES'],
        'eyeglasses': ['SUNGLASSES', 'EYEGLASSES', 'BLUE LIGHT GLASSES'],
        'handbag': ['HANDBAGS'],
        'bag': ['HANDBAGS'],
        'shoes': ['SHOES'],
        'jewelry': ['JEWELRY', 'COSTUME JEWELRY', 'COCO CRUSH', 'RINGS', 'NECKLACES', 'EARRINGS', 'BRACELETS'],
        'necklace': ['NECKLACES', 'JEWELRY'],
        'ring': ['RINGS', 'JEWELRY'],
        'earrings': ['EARRINGS', 'JEWELRY'],
        'bracelet': ['BRACELETS', 'JEWELRY'],
        'lipstick': ['LIPS'],
        'makeup': ['FACE', 'EYES', 'LIPS'],
        'perfume': ['FRAGRANCE', 'WOMEN', 'MEN'],
        'fragrance': ['FRAGRANCE', 'WOMEN', 'MEN'],
        'clothing': ['READY-TO-WEAR'],
        'dress': ['READY-TO-WEAR'],
        'wallet': ['SMALL LEATHER GOODS'],
        'belt': ['BELTS'],
        'scarf': ['SCARVES'],
    }

    try:
        image_rgb = image.convert('RGB')

        text_prompts = [f"a photo of a {label}" for label in category_labels.keys()]

        inputs = clip_processor(
            text=text_prompts,
            images=image_rgb,
            return_tensors="pt",
            padding=True
        ).to(clip_device)

        with torch.no_grad():
            outputs = clip_model(**inputs)
            logits_per_image = outputs.logits_per_image
            probs = logits_per_image.softmax(dim=1)

        max_prob_idx = probs.argmax().item()
        confidence = probs[0][max_prob_idx].item()

        detected_label = list(category_labels.keys())[max_prob_idx]
        detected_categories = category_labels[detected_label]

        return detected_categories, confidence, detected_label

    except Exception as e:
        st.warning(f"Détection catégorie échouée: {e}")
        return None, 0.0, None

# ============================================================================
# RECHERCHE OPTIMISÉE AVEC FILTRAGE INTELLIGENT
# ============================================================================

def advanced_image_search(query_embedding, embeddings_db, query_image=None, top_k=9):
    '''
    Recherche visuelle optimisée avec détection automatique de catégorie.
    1. Identifie la catégorie du produit dans l'image
    2. Filtre les résultats par cette catégorie
    3. Applique un seuil de similarité dynamique
    '''
    similarities = cosine_similarity([query_embedding], embeddings_db)[0]

    valid_mask = np.any(embeddings_db != 0, axis=1)
    similarities[~valid_mask] = -1

    detected_categories = None
    detected_label = None
    confidence = 0.0

    if query_image is not None:
        detected_categories, confidence, detected_label = detect_category_from_image(query_image)

        if detected_categories and confidence > 0.15:
            category_mask = df['category2_code'].isin(detected_categories).values
            similarities[~category_mask] = -1

    sorted_indices = np.argsort(similarities)[::-1]

    top_scores = similarities[sorted_indices[:top_k * 3]]
    valid_scores = top_scores[top_scores > 0]

    if len(valid_scores) > 0:
        mean_score = np.mean(valid_scores)
        std_score = np.std(valid_scores)
        threshold = max(0.65, mean_score - std_score)
    else:
        threshold = 0.65

    filtered_indices = []
    filtered_scores = []

    for idx in sorted_indices:
        if similarities[idx] >= threshold and len(filtered_indices) < top_k:
            filtered_indices.append(idx)
            filtered_scores.append(similarities[idx])

    if len(filtered_indices) < 3:
        filtered_indices = sorted_indices[:top_k].tolist()
        filtered_scores = similarities[sorted_indices[:top_k]].tolist()

    return np.array(filtered_indices), np.array(filtered_scores), detected_label, confidence

def precise_search(query_embedding, embeddings_db, query_text=None, top_k=9, use_filter=True):
    similarities = cosine_similarity([query_embedding], embeddings_db)[0]

    if use_filter and query_text:
        detected_cat = detect_category_from_query(query_text)
        if detected_cat:
            mask = df['category2_code'].isin(detected_cat).values
            if mask.any():
                similarities[~mask] = -1

    valid_idx = np.where(similarities > -1)[0]
    if len(valid_idx) == 0:
        return [], [], None

    sorted_idx = valid_idx[np.argsort(similarities[valid_idx])[::-1]]
    top_idx = sorted_idx[:min(top_k, len(sorted_idx))]

    return top_idx, similarities[top_idx], None

# ============================================================================
# AFFICHAGE
# ============================================================================

def show_results(indices, scores):
    if len(indices) == 0:
        st.error("❌ Aucun résultat")
        return

    st.markdown(f"### 🎯 Top {len(indices)} Résultats")
    cols = st.columns(3)

    for i, (idx, score) in enumerate(zip(indices, scores)):
        product = df.iloc[idx]
        with cols[i % 3]:
            try:
                if pd.notna(product.get('imageurl')):
                    img = Image.open(BytesIO(requests.get(product['imageurl'], timeout=5).content))
                    st.image(img, use_container_width=True)
            except:
                st.image("https://via.placeholder.com/300?text=No+Image", use_container_width=True)

            st.markdown(f"**{product.get('title', 'N/A')[:55]}...**")

            if pd.notna(product.get('category2_code')):
                st.markdown(f'<span class="category-badge">{product["category2_code"]}</span>',
                           unsafe_allow_html=True)

            if pd.notna(product.get('price')):
                st.markdown(f"💰 **{product['price']}€**")

            st.markdown(f'<div class="similarity-score">{score*100:.1f}%</div>',
                       unsafe_allow_html=True)

# ============================================================================
# INTERFACE
# ============================================================================

st.markdown('<h1 style="text-align:center;">✨ CHANEL Product Recommender</h1>',
            unsafe_allow_html=True)

with st.sidebar:
    st.markdown("## ⚙️ Configuration")
    mode = st.radio("Mode", ["📝 Texte", "🖼️ Image", "🔀 Multimodal"])
    top_k = st.slider("Résultats", 3, 6, 10)
    use_filter = st.checkbox("Filtrage catégorie", value=True)

    if mode == "🔀 Multimodal":
        alpha = st.slider("Poids Image/Texte", 0.0, 1.0, 0.5, 0.1)

st.markdown("---")

# MODE TEXTE
if mode == "📝 Texte":
    st.markdown("## 📝 Recherche par Texte")
    st.info("💡 Ex: Lunettes, Sac à main noir, Rouge à lèvres, Montre, Parfums, Maquillage, Bijoux, Vêtements, Colliers, Boucles d'oreilles")

    text_query = st.text_area("Description", placeholder="Décrivez le produit...", height=100)

    if text_query and st.button("🔍 Rechercher", use_container_width=True):
        with st.spinner("Recherche..."):
            text_de = translator.translate_to_german(text_query)
            st.info(f"🇫🇷 {text_query} → 🇩🇪 {text_de}")

            query_emb = extract_text_embedding(text_de)
            indices, scores, _ = precise_search(query_emb, text_emb, text_query, top_k, use_filter)

            st.markdown("---")
            show_results(indices, scores)

# MODE IMAGE OPTIMISÉ
elif mode == "🖼️ Image":
    st.markdown("## 🖼️ Recherche par Image")
    st.info("💡 Uploadez une photo de produit Chanel")

    uploaded = st.file_uploader("📤 Image", type=['png','jpg','jpeg'])

    if uploaded:
        col1, col2 = st.columns([1, 2])

        with col1:
            image = Image.open(uploaded)
            st.image(image, caption="Votre image", use_container_width=True)

        with col2:
            if st.button("🔍 Trouver similaires", use_container_width=True):
                with st.spinner("Analyse image..."):
                    query_emb = extract_image_embedding(image)
                    indices, scores, detected_label, confidence = advanced_image_search(
                        query_emb, visual_emb, query_image=image, top_k=top_k
                    )

                    if detected_label:
                        st.success(f"🎯 Produit détecté: **{detected_label.upper()}** (confiance: {confidence*100:.1f}%)")

                    st.markdown("---")
                    show_results(indices, scores)

# MODE MULTIMODAL
else:
    st.markdown("## 🔀 Recherche Multimodale")

    col1, col2 = st.columns(2)

    with col1:
        st.markdown("### 🖼️ Image")
        uploaded = st.file_uploader("📤 Image", type=['png','jpg','jpeg'])
        if uploaded:
            image = Image.open(uploaded)
            st.image(image, use_container_width=True)

    with col2:
        st.markdown("### 📝 Description")
        text_query = st.text_area("Style, couleur...", height=150)

    if uploaded and text_query and st.button("🔍 Recherche", use_container_width=True):
        with st.spinner("Analyse..."):
            text_de = translator.translate_to_german(text_query)
            st.info(f"🌐 {text_query} → {text_de}")

            img_emb = extract_image_embedding(image)
            txt_emb = extract_text_embedding(text_de)

            vis_sims = cosine_similarity([img_emb], visual_emb)[0]
            txt_sims = cosine_similarity([txt_emb], text_emb)[0]
            combined = alpha * vis_sims + (1-alpha) * txt_sims

            if use_filter:
                detected_cat = detect_category_from_query(text_query)
                if detected_cat:
                    mask = df['category2_code'].isin(detected_cat).values
                    combined[~mask] = -1

            valid = np.where(combined > -1)[0]
            top_idx = valid[np.argsort(combined[valid])[::-1]][:top_k]

            st.markdown("---")
            show_results(top_idx, combined[top_idx])

st.markdown(\"\"\"<br><br>
<div style='text-align:center; padding:20px; background:#f0f0f0; border-radius:10px;'>
    <p style='color:#666; font-weight:bold;'>🎯 Recherche PRÉCISE par IA</p>
    <p style='color:#999; font-size:13px;'>Créé par <b>ABDESLEM ISSAADI</b> —
    <a href='https://www.linkedin.com/in/abdeslem-issaadi-8aa6b0180/' target='_blank' style='color:#2196F3;'>LinkedIn</a></p>
</div>
\"\"\", unsafe_allow_html=True)
"""

with open('/content/app.py', 'w', encoding='utf-8') as f:
    f.write(app_code)

print("✅ Application optimisée créée!")

# ============================================================================
# CELLULE 7 : Lancement
# ============================================================================
import subprocess
import time

print("🚀 Lancement Streamlit...")

!pkill streamlit

process = subprocess.Popen(
    ['streamlit', 'run', '/content/app.py', '--server.port', '8501'],
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE
)

time.sleep(10)

public_url = ngrok.connect(8501)

print("\n" + "="*70)
print("✅ APPLICATION LANCÉE!")
print("="*70)
print(f"\n🔗 URL : {public_url}")
print("\n🎯 AMÉLIORATIONS RECHERCHE IMAGE :")
print("   ✓ Détection automatique de catégorie par CLIP")
print("   ✓ Filtrage intelligent par catégorie détectée")
print("   ✓ Affichage du type de produit identifié")
print("   ✓ Élimination des résultats hors catégorie")
print("\n⚠️ Gardez ce notebook ouvert")
print("="*70)
```
